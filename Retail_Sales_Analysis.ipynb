```python
import pandas as pd
import matplotlib.pyplot as plt

class RetailSalesAnalyzer:
    def __init__(self):
        self.data = pd.read_csv('retail_sales.csv')  # Replace 'your_file.csv' with your actual file name
        self.data['date'] = pd.to_datetime(self.data['Date'])  # Corrected 'to_datatime' to 'to_datetime'

    def data_clean(self):
        self.data.dropna(inplace=True)

    def total_sales_per_product(self):
        return self.data.groupby('Product')['Sales'].sum()

    def best_selling_product(self):
        return self.total_sales_per_product().sort_values(ascending=False).index[0]

    def average_daily_sales(self):
        return self.data['Sales'].mean()  # Corrected 'sales' to 'Sales'

    def plot_sales_trend(self):
        self.data.groupby('date')['Sales'].sum().plot(kind='line')  # Added missing parentheses
        plt.title('Sales Trend over Time')  # Corrected 'sales' to 'Sales'
        plt.xlabel('Date')  # Corrected 'xLable' to 'xlabel'
        plt.ylabel('Sales')  # Corrected 'yLabel' to 'ylabel'
        plt.show()

    def plot_sales_per_product(self):
        self.total_sales_per_product().plot(kind='bar')
        plt.title('Sales per Product')  # Corrected 'titles' to 'title'
        plt.xlabel('Product')  # Corrected 'xLable' to 'xlabel'
        plt.ylabel('Sales')  # Corrected 'yLabel' to 'ylabel'
        plt.show()

# Example usage
analyzer = RetailSalesAnalyzer()
analyzer.data_clean()
print('Total Sales per Product:\n', analyzer.total_sales_per_product())
print('Best Selling Product:', analyzer.best_selling_product())
print('Average Daily Sales:', analyzer.average_daily_sales())
analyzer.plot_sales_trend()
analyzer.plot_sales_per_product()
```
